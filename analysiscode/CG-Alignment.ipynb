```python
import MDAnalysis as mda
from numpy import *
import os
from pylab import *
import MDAnalysis.analysis.distances
import MDAnalysis.analysis.rms
from MDAnalysis.analysis import align
import glob
import scipy.stats
import sklearn
import sklearn.decomposition
import sklearn.preprocessing
import pandas as pd
import seaborn as sns
```

```python
cg_trimer = mda.Universe("../pdbs/trim5-public-master-CG-paper-pdbs/cg_L43_CA_tiled-SelectedTRIMER2.pdb")
cg_trim = mda.Universe("../pdbs/trim5-public-master-CG-paper-pdbs/cg_L43_dimer_tiled-SelectedTRIM2.pdb")
cg_trim.atoms.positions *=5.84
cg_trimer.atoms.positions *=5.84
cg_trim.atoms.write("cg_trim_expanded.pdb")
cg_trimer.atoms.write("cg_trimer_expanded.pdb")

aa_trimer = mda.Universe("../pdbs/hivcapsid/trimer.pdb")
aa_trim   = mda.Universe("../pdbs/2lm3_1.pdb")
```

```python
aa_trimPath = "state_dbscan_outputs_rhesus_3loopstate_1_trj0_frame65.pdb"
aa_trim = mda.Universe("PDBs_to_Dock/PCA_and_DBSCAN-on-all-3-loop-CA-distances/"+aa_trimPath)

cg_trim_resids=cg_trim.residues.resids.astype(str)
resid_str = "resid "
for i in range(len(cg_trim_resids)):
    resid_str+=cg_trim_resids[i]+" or resid "
resid_str = resid_str[:-10]



mobile0 = aa_trim.select_atoms("name CA and ("+resid_str+")").positions - aa_trim.select_atoms("name CA and ("+resid_str+")").center_of_mass()
ref0 = cg_trim.select_atoms("resid 290:495 and name CA and ("+resid_str+")").positions - cg_trim.select_atoms("resid 290:495 and name CA and ("+resid_str+")").center_of_mass()
R, rmsd = align.rotation_matrix(mobile0, ref0)



aa_trim.atoms.translate(-aa_trim.select_atoms("resid 290:495 and name CA and ("+resid_str+")").center_of_mass())
aa_trim.atoms.rotate(R)
aa_trim.atoms.translate(cg_trim.select_atoms("resid 290:495 and name CA and ("+resid_str+")").center_of_mass())

aa_trim.atoms.write("PDBs_to_Dock/PCA_and_DBSCAN-on-all-3-loop-CA-distances/CG_ALIGNED/"+aa_trimPath+"_aligned.pdb")
```
